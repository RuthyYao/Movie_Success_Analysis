# Overview

The movie industry is recovering from the Covid-19 pandemic with the cinema attendance picking up steadily and the box-office chasing the high of the pre-pandemic level. Miscrosoft,entrenched in the gaming and entertainemnt business, is looking to step into studio and movie production. Descriptive analysis of movie popularity score, box office performance and gross profits shows that Animation, Sci-fi and Adventure are the most popular and profitable genres with school holiday season (May-July and November/December) the best release time. From a financial perspective, an avereage annual budget of ~$500m is required to run a studio based on data from the current 12 major studios in the world. Miscosoft can use the insights to prepare the launch of the first movie, develop a solid grand budget as well as a detailed annual production plan. 

## Business Problem

To launch a new movie studio successfully, Microsoft has to find out the insights on the following important issues:

* What type of movie is the best performing?   
* How much budget does it need to produce for the selected genre and on average how much funding does it need to run a studio each year?
* What is the best months to release a movie?

Using the data on genres in conjuction with the movies popularity score and gross revenue, as well as production budget on studio and release date of over 2000 movies from the well-known movie databases - IMDB, Box Office Mojo, The Numbers and The Movie DataBase, I identify the top performing genres, the best release months and the annual budget to run a studio.

## Data Preparation

I took the following steps to prepare the data: 

1. Data cleaning. This invovles converting the data type, removing the irrelevant columns, and dealing with missing values

 
2. Merge/join the datatables to consoldiate the variables into one dataframe. I will choose the appropriate merge type to consolidate the data.


3. Feature engieering. This involves creating new features/columnbs to prepare for the data analysis.

### Data Cleaning

```python
#Subset the title_basics datatable to keep the relevant columns only.
title_basics = title_basics[['primary_title','genres']]
```

```python
title_basics.head()
```

```python
#Subset the tmdb datatable to keep the relevant columns only.
tmdb_movies = tmdb_movies[['title','release_date','vote_average','popularity']]
```

```python
tmdb_movies['release_date'] = pd.to_datetime(tmdb_movies['release_date'])
```

```python
tmdb_movies.head()
```

```python
tmdb_movies.info()
```

```python
#Remove the irrelevant columns in the TN's database.
tn_movie_gross =tn_movie_gross.drop(columns = ['id','domestic_gross','release_date'])
```

```python
tn_movie_gross.head()
```

```python
# Convert the budget and gross revenue columns to numerical.
tn_movie_gross[tn_movie_gross.columns[-2:]] = tn_movie_gross[tn_movie_gross.columns[-2:]].replace('[\$,]','',regex=True).astype(np.int64)
```

```python
tn_movie_gross.info()
```

```python
# subset the BOM's datatable to keep the title and studio only.
bom_movie_gross = bom_movie_gross[['title','studio']]
```

```python
bom_movie_gross.head()
```

### Merge the four dataframes

```python
# Merge the title_basics with the TMDB's dataframe to combine the genres with the popularity, vote score. 
genre_popularity = pd.merge(tmdb_movies,title_basics, left_on = 'title',right_on = 'primary_title', how = 'left').drop(columns = 'primary_title')
```

```python
genre_popularity.head()
```

```python
# Merge the genre_popularity dataframe with the TN's movie_gross dataframe to add the gross revenue and budget into the data.
genre_popularity_rev = pd.merge(genre_popularity,tn_movie_gross,left_on = 'title',right_on = 'movie', how = 'inner').drop(columns=['movie'])
```

```python
genre_popularity_rev.head()
```

```python
genre_popularity_rev.info()
```

```python
# Merge the above dataframe with the BOM studio data.
df_movie = pd.merge(genre_popularity_rev,bom_movie_gross,on = 'title',how = 'inner')
```

```python
df_movie.head()
```

```python
df_movie.info()
```

```python
# remove the rows that genres is missing
df_movie = df_movie.dropna()
```

```python
df_movie.info()
```

### Feature Engineering

I create a `profit` feature to analysis the profit a movie make. as a rule of thumb, studios get 50% of the gross revenue with the rest goes to the cinema. Hence I use 50% of the worldwide gross revenue deducted by the production budget calculate the profit.

```python
df_movie['profit'] = df_movie['worldwide_gross']*0.5-df_movie['production_budget']
```

I extract the `year` and `month`from the `release_date` columns.

```python
df_movie['release_year'] = df_movie['release_date'].dt.year
df_movie['release_month'] = df_movie['release_date'].dt.month
#df_movie['release_month'] = df_movie['release_date'].map(lambda x : x[5:7]).astype(np.int64)
```

I create `year_month` features for aggregating values by month in my analysis.

```python
df_movie['year_month'] = df_movie['release_date'].dt.strftime('%Y-%m')
```

I explode the `genres` into multiple rows with one genre in each row.

```python
df_movie_explode = df_movie.copy()
```

```python
# split the genres into a list
df_movie_explode['genres']=df_movie['genres'].str.split(pat=',',expand = False)
```

```python
# explode the genres list into multiple rows
df_movie_explode=df_movie_explode.explode('genres').dropna()
```

```python
df_movie_explode.info()
```

## Data Analysis

### The most popular genres

The `popularity` score from The Movie DataBase is built on the number of votes and views, as well as the number of users who mark it as "favorite" or save it in "watchlist". Hence I'll use this popularity score in conjuction with the box office gross revenue as the indicators of movie's popularity.  

```python
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(12,7))
for genres in df_movie_explode['genres'].unique():
    subset_df = df_movie_explode[df_movie_explode['genres'] == genres]
    x = subset_df['vote_average']
    y = subset_df['popularity']
    
    ax.scatter(x,y)

    
ax.set_xlabel('Average Vote Score')
ax.set_ylabel('Popularity')
ax.set_xticks(np.arange(0,11,1))
ax.set_title('Correlation between Popularity and vote score')
ax.legend(df_movie_explode['genres'].unique());
```

The scatter chart demonstrates that there is an correlation between the vote score and popularity and the correlation is stronger when the movie's vote score is between 4 and 8.5, which is also the range where the majority of the movies scored in. This add further confidence that we can use popularity score as a measure for the genre's success.

Now examine the average popularity and average worldwide gross revenue per genre by grouping the data by genres.

```python
genre_average = df_movie_explode.groupby(['genres'])[['popularity','worldwide_gross']].mean()
genre_average = genre_average.reset_index()
```

```python
genre_average
```

```python
fig, ax = plt.subplots()
x = genre_average['genres']
y = genre_average['popularity']

ax.set_xlabel('Average Popularity Score')
ax.set_ylabel('Genres')
ax.set_title('Average Popularity Score per Genre')

plt.barh(x,y);
```

Sci-fi has the highest score, followed by Adventure, Action, Fantasy and Amination. 

```python
fig, ax = plt.subplots()
x = genre_average['genres']
y = genre_average['worldwide_gross']

ax.set_xlabel('Average Gross Revenue')
ax.set_ylabel('Genres')
ax.set_title('Average Gross Revenue per Genre')

plt.barh(x,y);
```

Animations hit an average 3.9m gross revenue globally, the highest among all genres, followed by Adventure and Sci-fi. News and War are the worst performing genres.

Put together the popularity score and the worldwide gross revenue, we can conclude that Animation, Sci-fi and adventures are the the most successful movie genres. This is likely due to the broad audieance that those genres get. To the opposite, News and War are the underperforming genres.

### The most profitable genres

First examine the budget and profit from a per movie perspective.

```python
df_movie.describe()
```

Looking at the whole dataframe of ~2000 movies, on average, a movie can generate ~140m gross revenue on a production budget of 43m,returning ~27m profit to the studio. 

However, budget/profit may differ across the genres. Let's analyze the budget/profit by genres.

```python
# Group the data by genres to analyze the average production budget and gross profit for a movie per genre.
genre_average = df_movie_explode.groupby(['genres'])[['production_budget','profit']].mean()
genre_average = genre_average.reset_index()
```

```python
genre_average.sort_values(['production_budget','profit'], ascending =[False,False])
```

The data indicates that Adventure, Animation and Sci-fi, which are the most popular genres, also cost more to produce. Their production budget can reach the 100m mark. This could be driven by the sophistication of those movies that require significant investment in software, technology and high skilled labour. Nevertheless, those genres still make the most profit. 

```python
fig, ax = plt.subplots()
x = genre_average['genres']
y = genre_average['profit']

ax.set_xlabel('Average Gross Profit')
ax.set_ylabel('Genres')
ax.set_title('Average Gross Profit per Genre')

plt.barh(x,y);
```

The bar chart shows that Animation makes the highest profit, followed by Sci-fi, Adventure. This is aligned with our previous genre recommendation. Western, News and Wars are loss-making genres in general either due to low box office revenues or relatively high production cost. Those are the genres that Miscrsoft should avoid.  

### Annual Production Budget  to Run a Studio

```python
df_movie['release_year'].value_counts()
```

Movies before year 2010 are very sporatic. Those are old movies where the production budget are not quite relevant to the current age. Hence I decide to exclude them from the dataset for the yearly budget analysis.  

```python
# remove the movies that are released before 2010.
df_movie_filter = df_movie.loc[df_movie['release_year']>=2010]
```

```python
df_movie_filter.head()
```

I'll pivot the datatable by years and by studio.

```python
studio_movie_count = pd.pivot_table(df_movie_filter,index = 'studio', columns = 'release_year', values='title',aggfunc= 'count')
```

```python
studio_movie_count
```

There are 87 studios in toatl where most studios have a lot of missing values. Since I'm only interested in the top studios, I'll remove the studios with missing values.

Let's have a look how many movies are produced each year for a studio.

```python
studio_movie_count = studio_movie_count.dropna()
```

```python
studio_movie_count
```

```python
studio_movie_count.describe()
```

The movie production varies among the 12 studios. However, by looking at the mean and median of the movie production in the recent couple of years, we can safety say that on average, a studio produce 7-10 movies each year. 

Let's have a look of the annual production budget of each studio.

```python
studio_bud = pd.pivot_table(df_movie_filter,index = 'studio', columns = 'release_year', values='production_budget',aggfunc= 'sum').dropna()
```

```python
studio_bud
```

```python
year_moviecount = df_movie_filter[['release_year','title']].groupby('release_year').count()
year_bud = df_movie_filter[['release_year','production_budget']].groupby('release_year').sum()
```

```python
df_moviecount = pd.DataFrame(year_moviecount).rename(columns={'title':'num_titles'})
df_bud = pd.DataFrame(year_bud)
```

```python
df_per_movie_bud = pd.merge(df_moviecount,df_bud,on='release_year', how='left')
```

```python
df_per_movie_bud['bud_per_title'] = df_per_movie_bud['production_budget']/df_per_movie_bud['num_titles']
```

```python
df_per_movie_bud
```

```python
df_studio_bud = pd.DataFrame({'annual_budget':studio_bud.mean(),'num_titles':studio_movie_count.mean()})
```

```python
df_studio_bud
```

```python
df_studio_bud.mean()
```

```python
fig, ax =plt.subplots()
width = 0.7
df_studio_bud['annual_budget'].plot()
df_studio_bud['num_titles'].plot(secondary_y = True)

ax = plt.gca()
plt.show()
```

```python
studio_bud.describe()
```

Annual production budget also varies between studios. On an average basis, we can safely say that a studio need to invest ~500m every year in movie production.

By looking at 12 studios of the number of movies they produced each year and the annual production budget in the last 9 years, it shows that a studio needs average ~$530m budget to produce 7-10 movies averagely each year.

### Best Month to Release Movies

I use the `df_movie_filter` dataset examine which month achieve the highest average box office gross in the last 10 years.

```python
month = df_movie_filter['year_month'].sort_values().unique()
```

```python
movie_count = df_movie_filter[['year_month','title']].groupby('year_month').count()
df_movie_count = pd.DataFrame(movie_count).rename(columns ={'title': 'Number of Movies'})
```

```python
fig,ax=plt.subplots(figsize=(15,5))
                             
ax.set_xlabel('Month')
ax.set_ylabel('Number of Movies')
ax.set_title('Number of Movie released by Month')

count_plt = ax.plot(df_movie_count)
plt.xticks(month[::3],rotation = 60)
                 
plt.grid()
plt.show()
```

```python
movie_gross = df_movie_filter[['year_month','worldwide_gross']].groupby('year_month').mean()
df_movie_gross = pd.DataFrame(movie_gross).rename(columns ={'worldwide_gross': 'Average Gross Revenue'})
```

```python
fig,ax=plt.subplots(figsize=(15,5))
ax.set_xlabel('Month')
ax.set_ylabel('Average Gross Revenue')
ax.set_title('Avereage Gross Revenue by Month')

count_plt = ax.plot(df_movie_gross)
plt.xticks(month[::3],rotation = 60)
plt.grid()
plt.show()
```

```python
# Grouping the data by month.
rev_month = df_movie_filter[['release_month','worldwide_gross']].groupby('release_month').mean()
rev_month = rev_month.reset_index()
```

```python
fig, ax = plt.subplots(figsize = (8,5))

x= rev_month['release_month']
y= rev_month['worldwide_gross']

ax.set_xlabel('Month of Movie Release')
ax.set_ylabel('Average Gross Revenue')
ax.set_title('Average Gross Revenue by Month')
ax.set_xticks(np.arange(1,13,1))

plt.bar(x,y);
```

```python
rev_month
```

The data indicates that July or broadly May to July is the best time to release movie, followed by November. This seasonality can be attributed to the school holiday.  Also, Academy Awards could be another factor - studios try to release the movies by end of the year in order to be eligible for the Academy Awards entry at he begining of the coming year.

## Conclusions
Three insights generated from the above analysis:
* The best genres to chose are a mix of Animation, Sci-fi and Adventures as it gets the largest viewers and hit the highest box office revenue, potentially due to the broad audiences it received. Those genres also make the highest profit - averagely above 70m, more than double the profit an average movie made. To the contrast, News, Wars and Western are the genres that should be avoided as they generally have limited audience, low box office revenue and are loss-making.
* While Animation, Sci-fi and Adventures generate the highest box office revenue, the production cost is also very high. Microsoft will need to consider how much funding or budget they are planning to invest in the stuio business. Looking at the current movie production market, a studio invests averagely 0.5m into their production budget that creates averagely 7-10 movies each year. Miscrosoft can use this as a benchmark to develop their annual production plan.
* Box Office performance shows a strong seasonality. Looking into the last ten years' data, May to July are the best period to release movies followed by November.This could primarily driven by the school holiday. On top of that, the Academy Awards could also have an impact on the timing of release. While there are various factiors that could influence the result, it is advisable that Miscrosoft leverage this seasonality insight to plan the movie production.

***
Limitation of the Analysis:
* There are other factors that will impact a movie's success. For exmaple, the cast, producers and directors are not covered in this analysis. The runtime is another factor that is not in the scope of this analysis. Finally, from a financial success perspective, the data in this analysis only consider the box office revenue and production budget. There are other costs, such as marketing, distribution as well as other sources of revenue, such as streaming, DVD, etc. should be considered for a full-fledged financial analysis.
